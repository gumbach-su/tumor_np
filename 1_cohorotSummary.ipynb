```python
# define the variables of interest to display
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

patient_ids = [
    1, 1, 1, # NP113
    2, 2, 2, # NP119
    3, 3, # NP128
    4, 4, 4, 4, # NP134
    5, # NP147
    6, # NP148
    7, # NP149
    8, 8, # NP150
    ]

age_list = [
    52, 52, 52, # NP113
    34, 34, 34, # NP119
    34, 34, # NP128
    31, 31, 31, 31, # NP134
    47, # NP147
    47, # NP148
    27, # NP149
    31, 31, # NP150
]

gender_list = [
    1, 1, 1, # NP113
    1, 1, 1, # NP119
    0, 0, # NP128
    1, 1, 1, 1, # NP134
    1, # NP147
    1, # NP148
    0, # NP149
    1, 1, # NP150
]

diagnosis_list = [
    3, 3, 3, # NP113
    2, 2, 2, # NP119
    3, 3, # NP128
    1, 1, 1, 1, # NP134
    3, # NP147
    1, # NP148
    2, # NP149
    3, 3, # NP150
] # 1 = epilepsy, 2 = cavernoma, 3 = tumor

grade_list = [
    4, 4, 4, # NP113
    0, 0, 0, # NP119
    2, 2, # NP128
    0, 0, 0, 0, # NP134
    4, # NP147
    0, # NP148
    0, # NP149
    2, 2, # NP150
] # 0 = epilepsy or cavernoma, 2-4 = tumor

path_list = [
    2, 2, 2, # NP113
    0, 0, 0, # NP119
    2, 2, # NP128
    0, 0, 0, 0, # NP134
    2, # NP147
    0, # NP148
    0, # NP149
    1, 1, # NP150
] # 0 = neither, 1 = oligo, 2 = astro

idh_mut_list = [
    0, 0, 0, # NP113
    0, 0, 0, # NP119
    1, 1, # NP128
    0, 0, 0, 0, # NP134
    0, # NP147
    0, # NP148
    0, # NP149
    1, 1, # NP150
] # 0 = n/a or wt 1 = IDH-mutant

location_list = [
    0, 0, 0, # NP113
    0, 0, 0, # NP119
    0, 1, # NP128
    0, 0, 0, 0, # NP134
    1, # NP147
    0, # NP148
    0, # NP149
    1, 0, # NP150
    ] # 0 = frontal, 1 = temporoparietal

yield_list = [
    204, 217, 151, # NP113
    104, 201, 269, # NP119
    29, 12, # NP128
    71, 93, 50, 60, # NP134
    61, # NP147
    31, # NP148
    32, # NP149
    15, 10, # NP150
]
print(np.mean(np.array(yield_list)))
# Get average yield for diagnosis = 1 or 2
diagnosis_arr = np.array(diagnosis_list)
yield_arr = np.array(yield_list)

mask_1_2 = (diagnosis_arr == 1) | (diagnosis_arr == 2)
avg_yield_1_2 = yield_arr[mask_1_2].mean()
print("Average yield for diagnosis = 1 or 2:", avg_yield_1_2)

# Get average yield for diagnosis = 3
mask_3 = (diagnosis_arr == 3)
avg_yield_3 = yield_arr[mask_3].mean()
print("Average yield for diagnosis = 3:", avg_yield_3)


# Create data matrix
data_lists = [
    patient_ids,
    age_list,
    gender_list,
    diagnosis_list,
    grade_list,
    path_list,
    idh_mut_list,
    insertion_ids,
    yield_list
]

list_names = [
    'Patient ID',
    'Age',
    'Gender',
    'Diagnosis',
    'Grade',
    'Pathology',
    'IDH Mutation',
    'Location',
    'Yield'
]

# Convert to numpy arrays
data_matrix = np.array(data_lists)
n_rows, n_cols = data_matrix.shape

# Create main grid plot - clean without colorbars
# Make grid even smaller while keeping text size the same
fig1, axes = plt.subplots(n_rows, 1, figsize=(n_cols * 0.2, n_rows * 0.2), gridspec_kw={'hspace': 0})

# Create white-to-black colormap
cmap = mcolors.LinearSegmentedColormap.from_list('white_to_black', ['white', 'black'], N=256)

# Plot each row
for i, (data, name) in enumerate(zip(data_lists, list_names)):
    ax = axes[i]
    
    # Normalize data to 0-1 range for this row
    data_array = np.array(data)
    data_min, data_max = data_array.min(), data_array.max()
    
    # Handle case where min == max (constant values)
    if data_max == data_min:
        normalized_data = np.ones_like(data_array) * 0.5  # Middle gray
    else:
        normalized_data = (data_array - data_min) / (data_max - data_min)
    
    # Create color array
    colors = cmap(normalized_data)
    
    # Plot as horizontal bar chart - make each cell square by adjusting bar width
    x_positions = np.arange(n_cols)
    # Make bars square by setting width = height (1.0)
    bars = ax.barh(0, [1.0] * n_cols, left=x_positions, height=1.0, 
                   color=colors, edgecolor='none')  # Remove individual cell borders
    
    # Set labels and title
    ax.set_ylabel(name, fontsize=12, rotation=0, ha='right', va='center')
    ax.set_xlim(0, n_cols)
    ax.set_ylim(-0.5, 0.5)
    ax.set_xticks([])
    ax.set_yticks([])
    
    # Remove all spines except for border
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

# Add border around the entire grid
for ax in axes:
    # Add border around each subplot
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(1.5)
        spine.set_color('black')

# Add single border around the entire grid
fig1.patch.set_linewidth(2)
fig1.patch.set_edgecolor('black')

plt.tight_layout()
plt.subplots_adjust(bottom=0.1)
plt.show()

# Create separate colorbar plot - all colorbars in one column
n_cols_cbar = 1  # All colorbars in one column
n_rows_cbar = n_rows  # One row per colorbar

# Make colorbar plot proportional to the main grid plot
# Main grid: figsize=(n_cols * 0.2, n_rows * 0.2) = (4.4, 1.8)
# Colorbar plot should have similar proportions
fig2, axes2 = plt.subplots(n_rows_cbar, n_cols_cbar, figsize=(1.8, n_rows_cbar * 0.2), 
                           gridspec_kw={'hspace': 0.4, 'wspace': 0.1})

# Flatten axes for easier indexing
if n_rows_cbar == 1:
    axes2 = axes2.reshape(1, -1)
axes2_flat = axes2.flatten()

# Create colorbars
for i, (data, name) in enumerate(zip(data_lists, list_names)):
    if i < len(axes2_flat):
        ax = axes2_flat[i]
        
        # Normalize data to 0-1 range for this row
        data_array = np.array(data)
        data_min, data_max = data_array.min(), data_array.max()
        
        # Create colorbar
        sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=data_min, vmax=data_max))
        sm.set_array([])
        
        cbar = plt.colorbar(sm, ax=ax, orientation='vertical', shrink=0.8, aspect=3.0)
        cbar.ax.tick_params(labelsize=8, pad=2)
        cbar.outline.set_linewidth(0.5)
        # Set only min and max tick marks to prevent overlapping
        cbar.set_ticks([data_min, data_max])
        
        # No titles on colorbars
        
        # Remove axes
        ax.set_xticks([])
        ax.set_yticks([])
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['left'].set_visible(False)

# Hide unused subplots
for i in range(len(data_lists), len(axes2_flat)):
    axes2_flat[i].set_visible(False)

plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.95, bottom=0.05)
plt.show()

print("📊 Grid Plot Complete!")
print("🔍 Main grid: Clean square cells without spacing issues")
print("🎨 Colorbar plot: 3 colorbars per column, separate from main grid")
print("📋 Each colorbar shows: Name, Min - Max values")

print("📊 Grid Plot Complete!")
print("🔍 Each row shows a different data type")
print("🎨 White = minimum value, Black = maximum value")
print("📋 Values are displayed on each bar")
```
